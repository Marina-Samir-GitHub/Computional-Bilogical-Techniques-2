```python
import networkx as nx
```

```python
g =nx.Graph()
```

```python
g.add_nodes_from(['a','b','c','d','e'])
```

```python
g.add_edges_from([('a','c'),('c','d'),('b','d'),('c','e')])
```

```python
nx.draw(g,with_labels=True,node_color='red',node_size=2000,font_size=20,font_color='white')
```

```python
g.nodes()
```

```python
g.edges()
```

```python
g.number_of_nodes()
```

```python
g.number_of_edges()
```

```python
for i in g.neighbors('a'):
    print(i)
```
